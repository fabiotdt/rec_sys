Numpy


- numpy.ndarray --> data structure for manipulating vectors and matrices in a computationally efficient way

```python
import numpy as np # np --> commonly used alias for numpy

# Create a 1D numpy array

a = np.array([1, 2, 3, 4, 5])

a
```

```python
# Create a 2D numpy array (matrix)

b = np.array([[1, 2, 3, 4],
              [4, 5, 6, 7],
              [7, 8, 9, 10]])

b
```

```python
print(f'shape of a: {a.shape}') # --> 1D array with 5 elements
print(f'number of element in a: {a.size}') # --> for 1D array, size and shape are same

print(f'shape of b {b.shape}') # --> 2D array with 3 rows and 3 columns 
print(f'number of element in b: {b.size}') # --> 12 elements in the 2D array
```

```python
# Usefull properties of np.arrays --> reshaping

c = b.flatten() # --> flattens the 2D array into a 1D array
d = b.reshape(2, 6) # --> reshapes the 2D array into a 2x4 matrix

print(f'array c: {c} \n')
print(f'array d: {d}')
```

Matplotlib

- library for data-visualisation purposes
- usefull for data exploration, feature engineering and model performance evaluation

```python
import matplotlib.pyplot as plt # plt --> commonly used alias for matplotlib.pyplot


x = np.linspace(0, 10, 10) #  --> creates an array of 10 equally space elements from 0 to 10
y = x ** 2 # --> square of the elements in the array

plt.plot(x, y) # --> plots the graph
plt.title('y = x^2') # --> sets the title of the graph
plt.xlabel('x') # --> sets the label for x-axis
plt.ylabel('y') # --> sets the label for y-axis
plt.grid() # --> displays the grid
plt.show() # --> displays the graph

```

```python
# Scatter plot --> ploting the same data in a scatter plot

plt.scatter(x, y) # --> plots the scatter plot
plt.title('Scatterplot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python

x = np.linspace(-10, 10, 200)
y = np.exp(-(x**2)/(2*1))    # --> sigma = 1
z = np.exp(-(x**2)/(2*10))   # --> sigma = 10


plt.grid()
plt.xlabel('x')                     # add a label to the x axis
plt.ylabel('y')                     # add a label to the y axis

plt.xticks(np.arange(-10, 11, 2))   # specify in which point to place a tick on the x axis
plt.yticks(np.arange(0, 2.2, 0.2))  # and on the y axis

# rs- stands for red, squared markers, solid line
# yd-- stands for yellow, diamond markers, dashed line
plt.plot(x, y, 'rs-', markevery=10, label='sigma=1')    # rs --> red "-" --> solid line
plt.plot(x, z, 'yd--', markevery=10, label='sigma=10')  # yd --> yellow "--" --> dashed line

plt.ylim(-0.05,1.05)             # set bottom and top limits for y axis
plt.legend()  
plt.show()    
```

### Data Exploration of the dataset

```python
## Data exploration --> finding outliers
# Box wisker plot

# Columns --> 'Duration', 'Date', 'Pulse', 'Maxpulse', 'Calories'

df.boxplot()
plt.show()
```

```python
# Feature distribution

df.plot(kind='hist', bins = 30, subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

```python
# Dealing with outliers 

# 1. Remove the outliers
# 2. Replace the outliers with the mean/median value

# Remove the outliers using IQR (Interquartile range)

df_no_date = df.copy().drop('Date', axis=1)

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

for column in df_no_date.columns:
    df_no_date = df_no_date[~((df_no_date[column] < (Q1[column] - 1.5 * IQR[column])) | (df_no_date[column] > (Q3[column] + 1.5 * IQR[column])))]

df_no_date.boxplot()
plt.show()
```

```python
# Remoove the outliers using Z-score

from scipy import stats

df_no_date = df.copy().drop('Date', axis=1)
z = np.abs(stats.zscore(df_no_date))
z_threshold = 3 # --> threshold value for z-score, 3st similar to 99.7% confidence interval
                                                # 2st similar to 95% confidence interval
                                                # 1st similar to 68% confidence interval

df_no_date = df_no_date[(z < z_threshold).all(axis=1)]

df_no_date.boxplot()
plt.show()
```
